```python
import numpy as np
import matplotlib.pyplot as plt
from pybaseball import bwar_bat
from team_colors import team_color_map  # Ensure your team_colors.py is in the same directory

def plot_batter_war(player_name):
    # Get WAR stats from Baseball Reference
    data = bwar_bat(return_all=True)

    # Filter data for the given player
    player_data = data[data['name_common'] == player_name]

    # Sort the data by year to ensure it's in chronological order
    player_data = player_data.sort_values(by='year_ID')

    # Extract the relevant data for plotting
    years = player_data['year_ID'].to_numpy()
    war_totals = player_data['WAR'].to_numpy()
    team_ids = player_data['team_ID'].to_numpy()

    # Plot the WAR totals for each season
    plt.figure(figsize=(10, 6))

    # Track peak WAR and corresponding year
    peak_year_index = np.argmax(war_totals)
    peak_year = years[peak_year_index]
    peak_war = war_totals[peak_year_index]

    # Plot dots and lines
    for i in range(len(years)):
        color = team_color_map.get(team_ids[i], "#000000")  # Default to black if team ID is not found
        plt.plot(years[i], war_totals[i], marker='o', linestyle='-', color=color)
        
        # Add WAR number labels directly above the dot, slightly higher for better visibility
        plt.text(years[i], war_totals[i] + 0.3, f'{war_totals[i]:.1f}', fontsize=9, ha='center')

        # Add subtle lines between each dot (except the last one)
        if i < len(years) - 1:
            plt.plot(years[i:i+2], war_totals[i:i+2], linestyle='--', color=color, alpha=0.3)  # Set alpha to 0.3 for faint lines

    # Highlight the peak year with a circle
    plt.scatter(peak_year, peak_war, color='gold', s=100, edgecolor='black', label='PEAK', zorder=5)

    # Add titles and labels
    plt.title(f"{player_name}'s WAR by Season", fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('bWAR', fontsize=12)

    # Add a legend with lines representing performance categories
    plt.axhline(y=8, color='purple', linestyle='-', label='8+ MVP', alpha=0.3)  # Set alpha for faintness
    plt.axhline(y=5, color='blue', linestyle='-', label='5+ All-Star', alpha=0.3)
    plt.axhline(y=2, color='green', linestyle='-', label='2+ Starter', alpha=0.3)
    plt.axhline(y=0, color='orange', linestyle='-', label='0-2 Bench', alpha=0.3)
    plt.axhline(y=-0.1, color='red', linestyle='-', label='< 0 Below Replacement', alpha=0.3)

    # Set y-ticks explicitly to increments of 2
    max_war = int(np.ceil(max(war_totals)))  # Get the maximum WAR value and round up
    plt.yticks(np.arange(0, max_war + 2, 2))  # Create ticks from 0 to max_war with step of 2

    # Position the legend outside the plot to the right
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=8, frameon=False)

    # Show grid and the plot
    plt.grid(True)
    plt.xticks(np.arange(min(years), max(years) + 1, 1), rotation=45)
    plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make room for the legend
    
    # Display the plot
    plt.show()

# Example usage
plot_batter_war('Barry Bonds')  # Correct way to call the function
```
